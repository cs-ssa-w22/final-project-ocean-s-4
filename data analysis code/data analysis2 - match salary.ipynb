# Mapping industry with salary

```python
import numpy as np
import pandas as pd
```

### 1. Mapping NAICS code with average salary

### I have uploaded the generated file "industry_code_with_salary.csv" to github. You can skip this part and directly read this csv file for the next step.

```python
# read data
salary = pd.read_excel('./oes_research_2020_allsectors.xlsx', sheet_name = 'state_industry_M2020')
```

```python
salary.head()
```

```python
# drop "A_MEAN" rows including "#" and "*"
useless_value = ['#', "*"]
salary = salary[salary.A_MEAN.isin(useless_value) == False]
```

```python
# covert "A_MEAN" into int
salary['A_MEAN'] = pd.to_numeric(salary['A_MEAN'])
```

```python
# extract 'NAICS' and 'A_MEAN'
salary = salary[['NAICS', 'A_MEAN']]
```

```python
# drop duplicated rows
salary = salary.drop_duplicates()
```

```python
# keep only NAICS and mean salary
map_code_salary = salary.groupby('NAICS').mean()
```

```python
map_code_salary.head()
```

```python
map_code_salary.to_csv("industry_code_with_salary.csv",index = True)
```

### 2. Mapping company name with average salary

```python
# set your school name
school_name = 'Harvard University'
```

```python
# please download this csv file from github
map_salary = pd.read_csv('./industry_code_with_salary.csv')
```

```python
# read your school profile file
individual_profile = pd.read_csv('./Harvard University_with_industry_code0.csv')
```

```python
# extract NAICS code from school profile
individual_profile['NAICS'] = individual_profile.current_company_industry.str.extract(r'(\d\d)',expand=False)
```

```python
# map NAICS with average salary
individual_profile = pd.merge(individual_profile, map_salary, on="NAICS", how='left')
```

```python
individual_profile.head()
```

```python
# write csv
filename = str(school_name)+"with_salary.csv"
individual_profile.to_csv(filename,index = True)
```
