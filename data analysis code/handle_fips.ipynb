```python
import pandas as pd
import numpy as np
```

```python
                  
df = pd.read_csv('../data/complete_data.csv') ##fips: np.nan not 'nan' in this file
```

```python
df['fips'] = df['fips'].astype(pd.Int64Dtype())
df['fips'] = df['fips'].astype(str)
```

```python
df.loc[df['fips'].str.len() == 4, ['fips']] = '0' + df['fips'][df['fips'].str.len() == 4]
```

```python
df.loc[df['fips']=='0<NA>', ['fips']]='nan'
```

```python
df.to_csv('complete.csv', index=False)
```
